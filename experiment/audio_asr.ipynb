```python
from transformers import CLIPModel, CLIPProcessor
import os
import moviepy.editor as mp
from langchain_text_splitters import RecursiveCharacterTextSplitter
import torch
device = "cuda" if torch.cuda.is_available() else "cpu"
print(device)
MODEL_CACHE_DIR = "../model_cache"
INPUT_VIDEO = "../This Integral Breaks Math.mp4"
```

```python
clip_path = os.path.join(MODEL_CACHE_DIR, "clip_model")
clip_model = CLIPModel.from_pretrained(clip_path, local_files_only= True).to(device)
clip_processor = CLIPProcessor.from_pretrained(clip_path, local_files_only=True)
```

```python
def audio_extraction(path: str):
    audio_file = '../outputs/audio_file.mp3'
    video = mp.VideoFileClip(path)
    audio = video.audio
    
    audio.write_audiofile(audio_file)
    
    audio.close()
    video.close()
    return audio_file

def audio_to_text(audio_path: str, model) -> str:
    segments, info = model.transcribe(audio_path, beam_size=5)
    text = ""
    for segment in segments:
        text+=segment.text
    return text

def chunk_text(text):
        
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size = 100,
        chunk_overlap = 20
    )
    chunked_text = text_splitter.split_text(text=text)
    return chunked_text
```

```python
audio_file = audio_extraction(path= INPUT_VIDEO)
```

```python
text = audio_to_text(audio_file, model= model )
```

```python
chunked_text = chunk_text(text=text)
```

```python
clip_inputs = clip_processor(text= chunked_text, return_tensors="pt", padding= True, ).to(device)
with torch.no_grad():
    text_features = clip_model.get_text_features(**clip_inputs)
    text_features = text_features.pooler_output
    text_features = text_features / text_features.norm(p=2, dim=-1, keepdim=True)
text_features = text_features.cpu().numpy()
```
